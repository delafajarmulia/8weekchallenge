# **Phone Book**
1. menambahkan kontak
2. menyimpan file
3. muat kontak dari file
4. buat fungsi tambahan (tambah, hapus, cari dan tampilkan kontak)

```python
class PhoneBook:
    def __init__(self, filename="kontak.txt"):
        self.filename = filename
        self.kontak = {}
        self.load_data()

    # Load data
    def load_data(self):
        try:
            with open(self.filename, "r") as file:
                for line in file:
                    line = line.strip()
                    if line:
                        nama, nomor = map(str.strip, line.split(","))
                        self.kontak[nama] = nomor
        except FileNotFoundError:
            self.kontak = {}

    # Simpan data
    def save_data(self):
        with open(self.filename, "w") as file:
            for nama, nomor in self.kontak.items():
                file.write(f"{nama},{nomor}\n")

    # Tambah data
    def tambah(self, nama, nomor):
        self.kontak[nama] = nomor
        self.save_data()
        print("âœ“ Kontak berhasil ditambahkan")

    # Hapus data
    def hapus(self, nama):
        if nama in self.kontak:
            del self.kontak[nama]
            self.save_data()
            print("âœ“ Kontak berhasil dihapus")
        else:
            print("âœ— Nama tidak ditemukan")

    # Cari data
    def cari(self, nama):
      nama_lower = nama.lower()
      found = False

      for key, nomor in self.kontak.items():
          if key.lower() == nama_lower:
              print(f"{key} : {nomor}")
              found = True
              break

      if not found:
          print("Kontak tidak ditemukan")


    # Tampil data
    def tampil(self):
        if not self.kontak:
            print("âœ— Belum ada kontak")
        else:
            print("\nðŸ“’ Daftar Kontak:")
            for nama, nomor in self.kontak.items():
                print(f"- {nama} : {nomor}")

def main():
    buku = PhoneBook()

    while True:
        print("\n=== MENU BUKU TELEPON ===")
        print("1. Tambah Kontak")
        print("2. Hapus Kontak")
        print("3. Cari Kontak")
        print("4. Tampilkan Semua")
        print("5. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            nama = input("Nama: ")
            nomor = input("Nomor: ")
            buku.tambah(nama, nomor)

        elif pilihan == "2":
            nama = input("Nama yang ingin dihapus: ")
            buku.hapus(nama)

        elif pilihan == "3":
            nama = input("Nama yang dicari: ")
            buku.cari(nama)

        elif pilihan == "4":
            buku.tampil()

        elif pilihan == "5":
            print("ðŸ‘‹ Keluarâ€¦")
            buku.save_data()
            break

        else:
            print("Pilihan tidak valid!")

main()
```

# **Menu Restoran**
1. Menyimpan data dalam Dictionary
2. Simpan file ke JSON
3. Menampilkan menu file saat run program
4. Fitur: tambah menu, update harga dan tampilkan menu

```python
class MenuRestoran:
    def __init__(self, filename="menu.txt"):
        self.filename = filename
        self.menu = {}
        self.load_data()

    # Load menu dari file
    def load_data(self):
        try:
            with open(self.filename, "r") as file:
                for line in file:
                    line = line.strip()
                    if line:
                        nama, harga = line.split(",")
                        self.menu[nama] = float(harga)
        except FileNotFoundError:
            self.menu = {}

    # Simpan menu ke file
    def save_data(self):
        with open(self.filename, "w") as file:
            for nama, harga in self.menu.items():
                file.write(f"{nama},{harga}\n")

    # Tambah menu baru
    def tambah_menu(self, nama, harga):
        self.menu[nama] = harga
        print("Menu berhasil ditambahkan.")

    # Update harga menu
    def update_harga(self, nama, harga_baru):
        nama_lower = nama.lower()
        found = False

        for key, nama in self.menu.items():
            if key.lower() == nama_lower:
                self.menu[key] = harga_baru
                print("Harga berhasil diperbarui")
                found = True
                break

        if not found:
            print("Menu tidak ditemukan.")

    # Tampilkan semua menu
    def tampil_menu(self):
        if not self.menu:
            print("Menu masih kosong.")
        else:
            print("\nDaftar Menu Restoran:")
            for nama, harga in self.menu.items():
                print(f"- {nama}: Rp{harga:,.0f}")


# Program utama
def main():
    resto = MenuRestoran()

    while True:
        print("\n=== MENU RESTORAN ===")
        print("1. Tambah Menu")
        print("2. Update Harga")
        print("3. Tampilkan Menu")
        print("4. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            nama = input("Nama makanan: ")
            harga = float(input("Harga: "))
            resto.tambah_menu(nama, harga)

        elif pilihan == "2":
            nama = input("Nama makanan yang ingin diupdate: ")
            harga_baru = float(input("Harga baru: "))
            resto.update_harga(nama, harga_baru)

        elif pilihan == "3":
            resto.tampil_menu()

        elif pilihan == "4":
            resto.save_data()
            print("Data disimpan. Keluar...")
            break

        else:
            print("Pilihan tidak valid!")


main()
```

# **Pelacak Pengeluaran**
1. Pengelompokkan pengeluaran dengan Dictionary
2. Simpan dalam CSV
3. Baca dan analisis dari file
4. Menampilkan pengeluaran berdasarkan kategori

```python
import csv

class ExpenseTracker:
    def __init__(self, filename="pengeluaran.csv"):
        self.filename = filename
        self.data = {}
        self.load_data()

    # Baca data dari file CSV
    def load_data(self):
        try:
            with open(self.filename, mode="r", newline="") as file:
                reader = csv.reader(file)
                for row in reader:
                    kategori, jumlah = row
                    kategori = kategori.lower()
                    jumlah = float(jumlah)
                    if kategori in self.data:
                        self.data[kategori] += jumlah
                    else:
                        self.data[kategori] = jumlah
        except FileNotFoundError:
            self.data = {}

    # Simpan data ke CSV
    def save_data(self):
        with open(self.filename, mode="w", newline="") as file:
            writer = csv.writer(file)
            for kategori, jumlah in self.data.items():
                writer.writerow([kategori, jumlah])

    # Tambah pengeluaran
    def tambah(self, kategori, jumlah):
        kategori = kategori.lower()
        jumlah = float(jumlah)

        if kategori in self.data:
            self.data[kategori] += jumlah
        else:
            self.data[kategori] = jumlah

        print("Pengeluaran berhasil ditambahkan!")

    # Tampilkan total per kategori
    def tampilkan(self):
        if not self.data:
            print("Belum ada data pengeluaran.")
        else:
            print("\n=== Total Pengeluaran per Kategori ===")
            for kategori, jumlah in self.data.items():
                print(f"- {kategori.capitalize()} : Rp {jumlah:.2f}")


def main():
    tracker = ExpenseTracker()

    while True:
        print("\n=== MENU PELACAK PENGELUARAN ===")
        print("1. Tambah Pengeluaran")
        print("2. Tampilkan Total Pengeluaran")
        print("3. Keluar dan Simpan")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            kategori = input("Kategori: ")
            jumlah = input("Jumlah (Rp): ")
            tracker.tambah(kategori, jumlah)

        elif pilihan == "2":
            tracker.tampilkan()

        elif pilihan == "3":
            tracker.save_data()
            print("Data tersimpan. Keluarâ€¦")
            break

        else:
            print("Pilihan tidak valid!")


main()
```

# **Sistem Perpustakaan**
1. Nested Dictionary untuk buku dan detailnya
2. Simpan data dalam bentuk JSON
3. Fungsi: tambah buku, pinjam buku, kembalikan buku dan cari buku
4. Persistensi data dalam file

```python
import json
import os

class Perpustakaan:
    def __init__(self, filename="perpustakaan.json"):
        self.filename = filename
        self.buku = {}
        self.load_data()

    # Load data dari file JSON (jika ada)
    def load_data(self):
        if os.path.exists(self.filename):
            with open(self.filename, "r") as file:
                self.buku = json.load(file)
        else:
            self.buku = {}

    # Simpan data ke JSON
    def save_data(self):
        with open(self.filename, "w") as file:
            json.dump(self.buku, file, indent=4)

    # Tambah buku baru
    def tambah_buku(self, kode, judul, penulis, stok):
        kode = kode.lower()
        if kode in self.buku:
            print("Kode buku sudah ada!")
        else:
            self.buku[kode] = {
                "judul": judul,
                "penulis": penulis,
                "stok": stok
            }
            print("Buku berhasil ditambahkan!")

    # Pinjam buku
    def pinjam_buku(self, kode):
        kode = kode.lower()
        if kode in self.buku and self.buku[kode]["stok"] > 0:
            self.buku[kode]["stok"] -= 1
            print("Buku berhasil dipinjam!")
        else:
            print("Buku tidak tersedia!")

    # Kembalikan buku
    def kembalikan_buku(self, kode):
        kode = kode.lower()
        if kode in self.buku:
            self.buku[kode]["stok"] += 1
            print("Buku berhasil dikembalikan!")
        else:
            print("Kode buku tidak ditemukan!")

    # Cari buku berdasarkan kode
    def cari_buku(self, kode):
        kode = kode.lower()
        if kode in self.buku:
            buku = self.buku[kode]
            print(f"\n=== Data Buku ===")
            print(f"Judul   : {buku['judul']}")
            print(f"Penulis : {buku['penulis']}")
            print(f"Stok    : {buku['stok']}")
        else:
            print("Buku tidak ditemukan!")


def main():
    perpustakaan = Perpustakaan()

    while True:
        print("\n=== SISTEM PERPUSTAKAAN ===")
        print("1. Tambah Buku")
        print("2. Pinjam Buku")
        print("3. Kembalikan Buku")
        print("4. Cari Buku")
        print("5. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            kode = input("Kode Buku: ")
            judul = input("Judul: ")
            penulis = input("Penulis: ")
            stok = int(input("Stok: "))
            perpustakaan.tambah_buku(kode, judul, penulis, stok)

        elif pilihan == "2":
            kode = input("Kode Buku: ")
            perpustakaan.pinjam_buku(kode)

        elif pilihan == "3":
            kode = input("Kode Buku: ")
            perpustakaan.kembalikan_buku(kode)

        elif pilihan == "4":
            kode = input("Kode Buku: ")
            perpustakaan.cari_buku(kode)

        elif pilihan == "5":
            perpustakaan.save_data()
            print("Data tersimpan. Keluar...")
            break

        else:
            print("Pilihan tidak valid!")

main()
```
